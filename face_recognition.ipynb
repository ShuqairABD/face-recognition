```python
import os
import shutil
import face_recognition
import matplotlib.pyplot as plt
from google.colab import files

# Создание каталога для хранения загруженных изображений
if not os.path.exists("uploaded_images"):
    os.makedirs("uploaded_images")

# Загрузка изображений
uploaded = files.upload()

# Перемещение загруженных изображений в созданный каталог
for filename in uploaded.keys():
    source_path = filename
    destination_path = os.path.join("uploaded_images", filename)
    shutil.move(source_path, destination_path)

# Создание каталога для хранения результатов
if not os.path.exists("result_images"):
    os.makedirs("result_images")

```

```python
from matplotlib.patches import Rectangle

# Проходим по загруженным изображениям
for filename in os.listdir("uploaded_images"):
    # Загружаем изображение
    image = face_recognition.load_image_file(os.path.join("uploaded_images", filename))

    # Обнаруживаем лица на изображении
    face_locations = face_recognition.face_locations(image)

    # Визуализируем обнаруженные лица
    plt.imshow(image)
    ax = plt.gca()
    for face_location in face_locations:
        top, right, bottom, left = face_location
        width = right - left
        height = bottom - top
        rect = Rectangle((left, top), width, height, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
    plt.title("Detected Faces")
    plt.show()

    # Сохраняем изображение с обнаруженными лицами
    result_image_path = os.path.join("result_images", f"{filename.split('.')[0]}_result.png")
    plt.savefig(result_image_path)
    print(f"Result saved to: {result_image_path}")
```
